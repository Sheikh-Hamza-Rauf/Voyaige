```python
import pandas as pd
import re
import numpy as np
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import torch
from transformers import BertTokenizer, BertModel

# Load the dataset
hotel_df = pd.read_csv('new_hotel_data.csv')

# Initialize BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

# List of facilities
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house',
                   'Pet Friendly', 'Airport shuttle', 'Terrace', 'Room Service', 'BBQ Facilities', 
                   '24-Hour Front Desk', 'Child-friendly', 'Kitchen', 'Microwave', 'Free parking', 
                   'Breakfast Buffet', 'Swimming Pool', 'Paid parking', 'Entire villa', 'Air conditioning', 
                   'Balcony', 'Spa', 'Business Center', 'Free breakfast', 'Currency Exchange', 
                   'Outdoor pool', 'Air Conditioning', 'Laundry', 'Golf Course', 'cafe', 'Indoor pool', 
                   'Parking', 'Garden', 'Patio', 'Fireplace', 'Restaurant', 'Smoking Rooms', 
                   'Babysitting Service', 'Heating', 'Paid Wi-Fi', 'Wheelchair accessible', 'Entire chalet', 
                   'Lift', 'Smoke-free', 'Entire cottage', 'Library', 'Hot tub', 'Beach access', 
                   'Family Rooms', 'Free Wi-Fi']

# Function to preprocess structured fields to align with chatbot intent recognition
def preprocess_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."

    # Mention available facilities
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."

    return description.strip()

# Example of how to use preprocess_data on hotel data
hotel_df['description'] = hotel_df.apply(preprocess_data, axis=1)


# Function to get BERT embeddings
def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Ensure it returns a 2D numpy array

# Create embeddings for the dataset using 'description' instead of 'cleaned_text'
embedding_matrix = np.vstack([get_bert_embeddings(text) for text in hotel_df['description']])

# Function to get the most similar entry based on the user input using BERT embeddings
def get_most_similar_entry(user_input):
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)  # Reshape to 2D
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    most_similar_index = np.argmax(similarities)
    return hotel_df.iloc[most_similar_index]

# Define Intents
INTENTS = {
    'greet': {
        'keywords': ['hello', 'hi', 'hey', 'good morning', 'good evening']
    },
    'goodbye': {
        'keywords': ['goodbye', 'bye', 'see you', 'thanks', 'thank you']
    },
    'best_hotel': {
        'keywords': ['best hotel', 'top hotel', '5 star hotel', 'highest rated hotel', 'best priced hotel']
    },
    'best_priced_hotel': {
        'keywords': ['cheapest hotel', 'lowest cost hotel', 'budget hotel']
    }
}

# Function to classify intent based on user input
def classify_intent(user_input):
    user_input_lower = user_input.lower()
    for intent, details in INTENTS.items():
        for keyword in details['keywords']:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                return intent
    return 'unknown'

# Function to generate a detailed response for hotels, restaurants, or attractions
def generate_detailed_response(entry):
    response = f"**{entry['description']}**."
    return response

# Function to get the best hotel in a specific city
def get_best_hotel_in_city(city):
    best_hotel = hotel_df[hotel_df['city'].str.lower() == city.lower()].nsmallest(1, 'price')
    if not best_hotel.empty:
        return best_hotel.iloc[0]
    return None

# Function to handle user input and provide responses
def chatbot_response(user_input):
    intent = classify_intent(user_input)
    
    if intent == 'greet':
        return "Hello! How can I assist you today?"
    elif intent == 'goodbye':
        return "Goodbye! Have a great day!"
    elif intent == 'best_hotel':
        city = re.search(r'in (\w+)', user_input)  # Extract city from user input
        if city:
            city_name = city.group(1)  # Use group() to extract the matched city
            best_hotel_entry = get_best_hotel_in_city(city_name)
            if best_hotel_entry is not None:
                return generate_detailed_response(best_hotel_entry)
            else:
                return f"Sorry, I couldn't find any hotels in {city_name}. Please check the city name and try again."
    
    elif intent == 'best_priced_hotel':
        return get_best_priced_hotel()
    
    # If intent is unknown, search for the most similar entry
    similar_entry = get_most_similar_entry(user_input)
    
    return generate_detailed_response(similar_entry)

# Example conversation loop
def chat():
    print("Welcome to the Hotel Information Chatbot!")
    print("You can ask about hotels, restaurants, and attractions.")
    print("Type 'exit' to end the conversation.\n")
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye! Have a great day!")
            break
        
        response = chatbot_response(user_input)
        print(f"Chatbot: {response}\n")

# Start the chatbot
if __name__ == "__main__":
    chat()
```

```python
import pandas as pd
import re
import numpy as np
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import torch
from transformers import BertTokenizer, BertModel

# Load hotel dataset
hotel_df = pd.read_csv('new_hotel_data.csv')
print(f"Dataset loaded. Number of rows: {len(hotel_df)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

# List of facilities
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house',
                   'Pet Friendly', 'Airport shuttle', 'Terrace', 'Room Service', 'BBQ Facilities', 
                   '24-Hour Front Desk', 'Child-friendly', 'Kitchen', 'Microwave', 'Free parking', 
                   'Breakfast Buffet', 'Swimming Pool', 'Paid parking', 'Entire villa', 'Air conditioning', 
                   'Balcony', 'Spa', 'Business Center', 'Free breakfast', 'Currency Exchange', 
                   'Outdoor pool', 'Air Conditioning', 'Laundry', 'Golf Course', 'cafe', 'Indoor pool', 
                   'Parking', 'Garden', 'Patio', 'Fireplace', 'Restaurant', 'Smoking Rooms', 
                   'Babysitting Service', 'Heating', 'Paid Wi-Fi', 'Wheelchair accessible', 'Entire chalet', 
                   'Lift', 'Smoke-free', 'Entire cottage', 'Library', 'Hot tub', 'Beach access', 
                   'Family Rooms', 'Free Wi-Fi']

# Function to preprocess structured fields to align with chatbot intent recognition
def preprocess_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."
    
    # Mention available facilities
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."
    
    return description.strip()

hotel_df['description'] = hotel_df.apply(preprocess_data, axis=1)

# Function to get BERT embeddings
def get_bert_embeddings(text):
    print(f"Generating BERT embeddings for: {text[:50]}...")  # Show part of the text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Ensure it returns a 2D numpy array
    print(f"Embedding generated of shape: {embedding.shape}")
    return embedding

# Create embeddings for the dataset using 'description'
print("Generating BERT embeddings for the dataset...")
embedding_matrix = np.vstack([get_bert_embeddings(text) for text in hotel_df['description']])
print(f"Embedding matrix created with shape: {embedding_matrix.shape}")

# Function to get the most similar entry based on the user input using BERT embeddings
def get_most_similar_entry(user_input):
    print(f"Finding most similar entry for user input: {user_input}")
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)  # Reshape to 2D
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    most_similar_index = np.argmax(similarities)
    print(f"Most similar index: {most_similar_index}")
    return hotel_df.iloc[most_similar_index]

# Define Intents
INTENTS = {
    'greet': {
        'keywords': ['hello', 'hi', 'hey', 'good morning', 'good evening']
    },
    'goodbye': {
        'keywords': ['goodbye', 'bye', 'see you', 'thanks', 'thank you']
    },
    'best_hotel': {
        'keywords': ['best hotel', 'top hotel', '5 star hotel', 'highest rated hotel', 'best priced hotel', 'top']
    },
    'cheapest_hotel': {
        'keywords': ['cheapest hotel', 'lowest cost hotel', 'budget hotel']
    },
    'all_cities': {
        'keywords': ['hotels in all cities', 'list of all cities', 'available cities']
    }
}

# Function to classify intent based on user input
def classify_intent(user_input):
    print(f"Classifying intent for user input: {user_input}")
    user_input_lower = user_input.lower()
    for intent, details in INTENTS.items():
        for keyword in details['keywords']:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                print(f"Intent '{intent}' matched with keyword '{keyword}'.")
                return intent
    print("No matching intent found.")
    return 'unknown'

# Function to generate a detailed response for hotels, restaurants, or attractions
def generate_detailed_response(entry):
    
    response = f"**{entry['description']}**."
    return response

# Function to get the best hotels in a specific city
def get_best_hotels_in_city(city, n=5):
    print(f"Searching for the top {n} hotels in city: {city}")
    best_hotels = hotel_df[hotel_df['city'].str.lower() == city.lower()].nlargest(n, 'price')
    if not best_hotels.empty:
        print(f"{n} best hotels found.")
        return best_hotels
    print(f"No hotels found for city: {city}")
    return None

# Function to get the cheapest hotel in a specific city
def get_cheapest_hotel_in_city(city):
    print(f"Searching for the cheapest hotel in city: {city}")
    cheapest_hotel = hotel_df[hotel_df['city'].str.lower() == city.lower()].nsmallest(1, 'price')
    if not cheapest_hotel.empty:
        print("Cheapest hotel found.")
        return cheapest_hotel.iloc[0]
    print(f"No hotel found for city: {city}")
    return None

# Function to get unique cities from the dataset
def get_unique_cities():
    print("Fetching list of unique cities.")
    unique_cities = hotel_df['city'].unique()
    print(f"Unique cities: {unique_cities}")
    return unique_cities

# Function to extract number of hotels from user input
def extract_number(user_input):
    numbers = re.findall(r'\d+', user_input)
    return int(numbers[0]) if numbers else 5  # Default to 5 if no number is found

# Function to handle user input and provide responses
def chatbot_response(user_input):
    print(f"Received user input: {user_input}")
    intent = classify_intent(user_input)
    
    if intent == 'greet':
        return "Hello! How can I assist you today?"
    elif intent == 'goodbye':
        return "Goodbye! Have a great day!"
    elif intent == 'best_hotel':
        city = re.search(r'in ([\w\s]+)', user_input)  # Extract city from user input
        n = extract_number(user_input)  # Extract the number of hotels requested
        if city:
            city_name = city.group(1)  # Use group() to extract the matched city
            best_hotels_entry = get_best_hotels_in_city(city_name, n)
            if best_hotels_entry is not None:
                response = f"Here are the top {n} hotels in {city_name}:\n"
                for _, row in best_hotels_entry.iterrows():
                    response += generate_detailed_response(row) + "\n"
                    print("=====================================================")
                return response.strip()
            else:
                return f"Sorry, I couldn't find any hotels in {city_name}. Please check the city name and try again."
    
    elif intent == 'cheapest_hotel':
        city = re.search(r'in ([\w\s]+)', user_input)  # Extract city from user input
        if city:
            city_name = city.group(1)  # Extract city name
            cheapest_hotel_entry = get_cheapest_hotel_in_city(city_name)
            if cheapest_hotel_entry is not None:
                return generate_detailed_response(cheapest_hotel_entry)
            else:
                return f"Sorry, I couldn't find any hotels in {city_name}. Please check the city name and try again."
    
    elif intent == 'all_cities':
        cities = get_unique_cities()
        return f"Here is a list of available cities: {', '.join(cities)}."
    
    # If intent is unknown, search for the most similar entry
    print("Attempting to find the most similar entry.")
    similar_entry = get_most_similar_entry(user_input)
    
    return generate_detailed_response(similar_entry)

# Example conversation loop
def chat():
    print("Welcome to the hotel chatbot! Type 'exit' to end the chat.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Chatbot: Goodbye! Have a great day!")
            break
        response = chatbot_response(user_input)
        print(f"Chatbot: {response}")

# Start chatting
if __name__ == "__main__":
    chat()
```

```python
import pandas as pd
import re
import numpy as np
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import torch
from transformers import BertTokenizer, BertModel

# Load hotel dataset
hotel_df = pd.read_csv('new_hotel_data.csv')
print(f"Dataset loaded. Number of rows: {len(hotel_df)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

# List of facilities
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house',
                   'Pet Friendly', 'Airport shuttle', 'Terrace', 'Room Service', 'BBQ Facilities', 
                   '24-Hour Front Desk', 'Child-friendly', 'Kitchen', 'Microwave', 'Free parking', 
                   'Breakfast Buffet', 'Swimming Pool', 'Paid parking', 'Entire villa', 'Air conditioning', 
                   'Balcony', 'Spa', 'Business Center', 'Free breakfast', 'Currency Exchange', 
                   'Outdoor pool', 'Air Conditioning', 'Laundry', 'Golf Course', 'cafe', 'Indoor pool', 
                   'Parking', 'Garden', 'Patio', 'Fireplace', 'Restaurant', 'Smoking Rooms', 
                   'Babysitting Service', 'Heating', 'Paid Wi-Fi', 'Wheelchair accessible', 'Entire chalet', 
                   'Lift', 'Smoke-free', 'Entire cottage', 'Library', 'Hot tub', 'Beach access', 
                   'Family Rooms', 'Free Wi-Fi']

# Function to preprocess structured fields to align with chatbot intent recognition
def preprocess_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."
    
    # Mention available facilities
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."
    
    return description.strip()

hotel_df['description'] = hotel_df.apply(preprocess_data, axis=1)

# Function to get BERT embeddings
def get_bert_embeddings(text):
    print(f"Generating BERT embeddings for: {text[:50]}...")  # Show part of the text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Ensure it returns a 2D numpy array
    print(f"Embedding generated of shape: {embedding.shape}")
    return embedding

# Create embeddings for the dataset using 'description'
print("Generating BERT embeddings for the dataset...")
embedding_matrix = np.vstack([get_bert_embeddings(text) for text in hotel_df['description']])
print(f"Embedding matrix created with shape: {embedding_matrix.shape}")

# Function to get the most similar entry based on the user input using BERT embeddings
def get_most_similar_entry(user_input):
    print(f"Finding most similar entry for user input: {user_input}")
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)  # Reshape to 2D
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    most_similar_index = np.argmax(similarities)
    print(f"Most similar index: {most_similar_index}")
    return hotel_df.iloc[most_similar_index]

# Define Intents
INTENTS = {
    'greet': {
        'keywords': ['hello', 'hi', 'hey', 'good morning', 'good evening']
    },
    'goodbye': {
        'keywords': ['goodbye', 'bye', 'see you', 'thanks', 'thank you']
    },
    'best_hotel': {
        'keywords': ['best hotel', 'top hotel', '5 star hotel', 'highest rated hotel', 'best priced hotel', 'top']
    },
    'cheapest_hotel': {
        'keywords': ['cheapest hotel', 'lowest cost hotel', 'budget hotel']
    },
    'all_cities': {
        'keywords': ['hotels in all cities', 'list of all cities', 'available cities','name all cities']
    }
}

# Function to classify intent based on user input
def classify_intent(user_input):
    print(f"Classifying intent for user input: {user_input}")
    user_input_lower = user_input.lower()
    for intent, details in INTENTS.items():
        for keyword in details['keywords']:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                print(f"Intent '{intent}' matched with keyword '{keyword}'.")
                return intent
    print("No matching intent found.")
    return 'unknown'

# Function to generate a detailed response for hotels, restaurants, or attractions
def generate_detailed_response(entry):
    print(f"Generating detailed response for entry: {entry['name']}")
    response = f"**{entry['description']}**."
    return response

# Function to get the best hotels in a specific city
def get_best_hotels_in_city(city, n=5):
    print(f"Searching for the top {n} hotels in city: {city}")
    best_hotels = hotel_df[hotel_df['city'].str.lower() == city.lower()].nlargest(n, 'price')
    if not best_hotels.empty:
        print(f"{n} best hotels found.")
        return best_hotels
    print(f"No hotels found for city: {city}")
    return None

# Function to get the cheapest hotel in a specific city
def get_cheapest_hotel_in_city(city):
    print(f"Searching for the cheapest hotel in city: {city}")
    cheapest_hotel = hotel_df[hotel_df['city'].str.lower() == city.lower()].nsmallest(1, 'price')
    if not cheapest_hotel.empty:
        print("Cheapest hotel found.")
        return cheapest_hotel.iloc[0]
    print(f"No hotel found for city: {city}")
    return None

# Function to get unique cities from the dataset
def get_unique_cities():
    print("Fetching list of unique cities.")
    unique_cities = hotel_df['city'].unique()
    print(f"Unique cities: {unique_cities}")
    return unique_cities

# Function to extract number of hotels from user input
def extract_number(user_input):
    numbers = re.findall(r'\d+', user_input)
    return int(numbers[0]) if numbers else 5  # Default to 5 if no number is found

# Function to handle user input and provide responses
def chatbot_response(user_input):
    print(f"Received user input: {user_input}")
    intent = classify_intent(user_input)
    
    if intent == 'greet':
        return "Hello! How can I assist you today?"
    elif intent == 'goodbye':
        return "Goodbye! Have a great day!"
    elif intent == 'best_hotel':
        city = re.search(r'in ([\w\s]+)', user_input)  # Extract city from user input
        n = extract_number(user_input)  # Extract the number of hotels requested
        if city:
            city_name = city.group(1)  # Use group() to extract the matched city
            best_hotels_entry = get_best_hotels_in_city(city_name, n)
            if best_hotels_entry is not None:
                response = f"Here are the top {n} hotels in {city_name}:\n"
                for _, row in best_hotels_entry.iterrows():
                    response += generate_detailed_response(row) + "\n"
                return response.strip()
            else:
                return f"Sorry, I couldn't find any hotels in {city_name}. Please check the city name and try again."
    
    elif intent == 'cheapest_hotel':
        city = re.search(r'in ([\w\s]+)', user_input)  # Extract city from user input
        if city:
            city_name = city.group(1)  # Extract city name
            cheapest_hotel_entry = get_cheapest_hotel_in_city(city_name)
            if cheapest_hotel_entry is not None:
                return generate_detailed_response(cheapest_hotel_entry)
            else:
                return f"Sorry, I couldn't find any hotels in {city_name}. Please check the city name and try again."
    
    elif intent == 'all_cities':
        cities = get_unique_cities()
        return f"Here is a list of available cities: {', '.join(cities)}."
    
    # If intent is unknown, search for the most similar entry
    print("Attempting to find the most similar entry.")
    similar_entry = get_most_similar_entry(user_input)
    
    return generate_detailed_response(similar_entry)

# Example conversation loop
def chat():
    print("Welcome to the hotel chatbot! Type 'exit' to end the chat.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Chatbot: Goodbye! Have a great day!")
            break
        response = chatbot_response(user_input)
        print(f"Chatbot: {response}")

# Start chatting
if __name__ == "__main__":
    chat()
```

```python
import pandas as pd
import re
import numpy as np
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import torch
import os
from transformers import BertTokenizer, BertModel

# Load datasets
hotel_df = pd.read_csv('new_hotel_data.csv')
ratings_df = pd.read_json('clean_hotel_review_data.json')
restaurant_df = pd.read_json('new_restaurant_db.restaurants_data.json')
attraction_df = pd.read_csv('attraction.csv')

#print(f"Datasets loaded. Number of hotels: {len(hotel_df)}, Number of restaurants: {len(restaurant_df)}, Number of attractions: {len(attraction_df)}")
# Merge hotel ratings with hotel data on 'hotel_id'
merged_df = hotel_df.merge(ratings_df, on='hotel_id', how='left')

# Calculate average rating for each hotel
average_ratings = merged_df.groupby('hotel_id')['rating'].mean().reset_index()
average_ratings.columns = ['hotel_id', 'average_rating']

# Merge the average ratings back into the hotel DataFrame
hotel_df = hotel_df.merge(average_ratings, on='hotel_id', how='left')

# Rename the rating column to avoid confusion
hotel_df.rename(columns={'average_rating': 'Rating'}, inplace=True)

restaurant_ratings_df = pd.read_json('new_restaurant_db.restaurants_reviews.json')
# Load restaurant ratings dataset
restaurant_merged_df = restaurant_df.merge(restaurant_ratings_df, on='restaurant_id', how='left')

# Calculate average rating for each hotel
average_restaurant_ratings = restaurant_merged_df.groupby('restaurant_id')['rating'].mean().reset_index()
average_restaurant_ratings.columns = ['restaurant_id', 'average_restaurant_ratings']

# Merge the average ratings back into the hotel DataFrame
restaurant_df = restaurant_df.merge(average_restaurant_ratings, on='restaurant_id', how='left')

# Rename the rating column to avoid confusion
restaurant_df.rename(columns={'average_restaurant_ratings': 'Rating'}, inplace=True)

restaurant_df['Rating'] = restaurant_df['Rating'].fillna(3.5)



tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

# List of facilities (as previously defined)
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house',
                   'Pet Friendly', 'Airport shuttle', 'Terrace', 'Room Service', 'BBQ Facilities', 
                   '24-Hour Front Desk', 'Child-friendly', 'Kitchen', 'Microwave', 'Free parking', 
                   'Breakfast Buffet', 'Swimming Pool', 'Paid parking', 'Entire villa', 'Air conditioning', 
                   'Balcony', 'Spa', 'Business Center', 'Free breakfast', 'Currency Exchange', 
                   'Outdoor pool', 'Air Conditioning', 'Laundry', 'Golf Course', 'cafe', 'Indoor pool', 
                   'Parking', 'Garden', 'Patio', 'Fireplace', 'Restaurant', 'Smoking Rooms', 
                   'Babysitting Service', 'Heating', 'Paid Wi-Fi', 'Wheelchair accessible', 'Entire chalet', 
                   'Lift', 'Smoke-free', 'Entire cottage', 'Library', 'Hot tub', 'Beach access', 
                   'Family Rooms', 'Free Wi-Fi']

# Function to preprocess structured fields for hotels
def preprocess_hotel_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."
    
    # Mention available facilities
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."
    
    return description.strip()

hotel_df['description'] = hotel_df.apply(preprocess_hotel_data, axis=1)

# Function to preprocess restaurant and attraction data
def preprocess_restaurant_data(row):
    return f"{row['name']} located at {row['address']} and its phone number is {row['phone_number']} in {row['city']}. The provided services are {row['service']} and opening hour is {row['Open_hour']}."

def preprocess_attraction_data(row):
    return f"{row['Name']} is a popular attraction located in {row['City']}. Its address is {row['Address']}, its category is {row['Category']}, its rating is {row['Rating']}, and its availability time is {row['Availability']}. Description: {row['Description']}."

restaurant_df['description'] = restaurant_df.apply(preprocess_restaurant_data, axis=1)
attraction_df['description'] = attraction_df.apply(preprocess_attraction_data, axis=1)


# Function to get BERT embeddings
def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Ensure it returns a 2D numpy array
    return embedding

# Function to save embeddings to a file if not already saved
def save_embeddings(data_frame, description_column, file_name):
    if not os.path.exists(file_name):
        print(f"Generating and saving embeddings to {file_name}...")
        embedding_matrix = np.vstack([get_bert_embeddings(text) for text in data_frame[description_column]])
        np.save(file_name, embedding_matrix)
        print(f"Embeddings saved to {file_name}.")
    else:
        print(f"Loading embeddings from {file_name}...")
        embedding_matrix = np.load(file_name)
        print(f"Loaded embeddings of shape {embedding_matrix.shape}.")
    return embedding_matrix

# Create/load embeddings for the dataset using 'description'
print("Generating or loading BERT embeddings for the datasets...")

# Hotel embeddings
hotel_embedding_matrix = save_embeddings(hotel_df, 'description', 'hotel_embeddings.npy')

# Restaurant embeddings
restaurant_embedding_matrix = save_embeddings(restaurant_df, 'description', 'restaurant_embeddings.npy')

# Attraction embeddings
attraction_embedding_matrix = save_embeddings(attraction_df, 'description', 'attraction_embeddings.npy')


# Function to get the most similar entry based on user input
def get_most_similar_entry(user_input, embedding_matrix, data_frame):
    print(f"Finding most similar entry for user input: {user_input}")
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)  # Reshape to 2D
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    most_similar_index = np.argmax(similarities)
    print(f"Most similar index: {most_similar_index}")
    return data_frame.iloc[most_similar_index]

# Define Intents
INTENTS = {
    'greet': {
        'keywords': ['hello', 'hi', 'hey', 'good morning', 'good evening']
    },
    'goodbye': {
        'keywords': ['goodbye', 'bye', 'see you', 'thanks', 'thank you']
    },
    'best_hotel': {
        'keywords': ['best hotel', 'top hotel', '5 star hotel', 'highest rated hotel', 'best priced hotel', 'top','hotel']
    },
    'cheapest_hotel': {
        'keywords': ['cheapest hotel', 'lowest cost hotel', 'budget hotel']
    },
    'all_cities': {
        'keywords': ['hotels in all cities', 'list of all cities', 'available cities']
    },
    'itinerary': {
        'keywords': ['plan an itinerary', 'itinerary', 'things to do', 'what to do in']
    },
    'update_itinerary': {
        'keywords': ['update itinerary', 'add to itinerary', 'modify itinerary', 'change itinerary']
    },
    'delete_from_itinerary': {
        'keywords': ['delete from itinerary', 'remove from itinerary', 'discard itinerary']
    },
    'best_restaurant': {
        'keywords': ['best restaurant', 'top restaurant', '5 star restaurant', 'highest rated restaurant', 'best priced restaurant','restaurants']
    },
    'best_attraction': {
        'keywords': ['best attraction', 'top attraction', 'must-see attraction','attractions']
    },
}

# Function to classify intent based on user input
def classify_intent(user_input):
    print(f"Classifying intent for user input: {user_input}")
    user_input_lower = user_input.lower()
    for intent, details in INTENTS.items():
        for keyword in details['keywords']:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                print(f"Intent '{intent}' matched with keyword '{keyword}'.")
                return intent
    print("No matching intent found.")
    return 'unknown'

# Function to generate a detailed response
def generate_detailed_response(entry, type='hotel'):
    if type == 'hotel':
        print('===================================================')
        response = f"**{entry['name']}**\nAddress: {entry['address']}\nPrice: Rs{entry['price']} per night\nAbout: {entry['about']}\nRating : {entry['Rating']}\nAvailable facilities: {', '.join([facility for facility in facilities_list if entry.get(facility, 0) == 1])}\n"
    elif type == 'restaurant':
        print('===================================================')
        response = f"**{entry['name']}**\nAddress: {entry['address']}\nPhone: {entry['phone_number']}\nServices: {entry['service']}\nOpening Hour: {entry['Open_hour']}\n Rating: {entry['Rating']}"
    elif type == 'attraction':
        print('===================================================')
        response = f"**{entry['Name']}**\nAddress: {entry['Address']}**\nCategory: {entry['Category']}\nRating: {entry['Rating']}\nAvailability: {entry['Availability']}\nDescription: {entry['Description']}\n"
    
    return response


# Function to get the best hotels in a specific city based on highest rating and price
def get_best_hotels_in_city(city, n=5):
    print(f"Searching for the top {n} hotels in city: {city} based on highest rating and price.")

    # Filter hotels by city
    city_hotels = hotel_df[hotel_df['city'].str.lower() == city.lower()]

    # Check if there are hotels in the specified city
    if not city_hotels.empty:
        # Sort by 'rating' (highest first) and 'price' (highest first)
        best_hotels = city_hotels.sort_values(by=['Rating', 'price'], ascending=[False, False]).head(n)

        print(f"{n} best hotels found based on highest rating and price.")
        return best_hotels

    # If no hotels are found, return None
    print(f"No hotels found for city: {city}")
    return None

# Function to get the cheapest hotel in a specific city
def get_cheapest_hotels_in_city(city, n=5):
    print(f"Searching for the top {n} cheapest hotels in city: {city}")

    # Filter hotels by city
    city_hotels = hotel_df[hotel_df['city'].str.lower() == city.lower()]

    # Check if there are hotels in the specified city
    if not city_hotels.empty:
        # Sort by 'rating' (highest first) and 'price' (lowest first)
        cheapest_hotels = city_hotels.sort_values(by=['Rating', 'price'], ascending=[False, True]).head(n)

        print(f"{n} cheapest hotels found based on highest rating and price.")
        return cheapest_hotels

    # If no hotels are found, return None
    print(f"No hotels found for city: {city}")
    return None
# Function to get the best attractions in a specific city
def get_best_attractions_in_city(city, n=5):
   
    city_attractions = attraction_df[attraction_df['City'].str.lower() == city.lower()]
    best_attractions = city_attractions.nlargest(n, 'Rating')  
    if not best_attractions.empty:
        print(f"{n} best attractions found.")
        return best_attractions
    print(f"No attractions found for city: {city}")
    return None



# Function to fetch the best restaurant in a specific city
def get_best_restaurant_in_city(city):
    print(f"Searching for the best restaurant in city: {city}")
    best_restaurant = restaurant_df[restaurant_df['city'].str.lower() == city.lower()].nlargest(5, 'Rating')
    if not best_restaurant.empty:
        print("Best restaurant found.")
        return best_restaurant  # Return the DataFrame instead of a single row
    print(f"No restaurant found for city: {city}")
    return None

def create_itinerary_for_city(city,budget):
    budget = float(budget)
    # Fetch 3 best hotels
    hotels_within_budget = hotel_df[(hotel_df['city'].str.lower() == city.lower()) & (hotel_df['price'] <= budget)]
    best_hotels = hotels_within_budget.nlargest(3, 'Rating')  # Get the top 3 hotels based on rating

    
    # Fetch 3 best restaurants
    best_restaurants = restaurant_df[restaurant_df['city'].str.lower() == city.lower()].nlargest(3, 'Rating')
    
    # Fetch 3 best attractions
    best_attractions = attraction_df[attraction_df['City'].str.lower() == city.lower()].nlargest(3, 'Rating')

    return best_hotels, best_restaurants, best_attractions


# Function to update itinerary
def update_itinerary(itinerary, new_activity):
    print(f"Updating itinerary by adding activity: {new_activity}")
    itinerary['activities'].append(new_activity)
    return itinerary

# Function to delete an activity from the itinerary
def delete_activity_from_itinerary(itinerary, activity):
    print(f"Deleting activity: {activity} from itinerary")
    if activity in itinerary['activities']:
        itinerary['activities'].remove(activity)
    return itinerary
    






# Chatbot conversation loop
def chatbot():
    user_itinerary = {'location': '', 'activities': []}  # Initialize user itinerary
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Chatbot: Goodbye! Have a great day!")
            break

        intent = classify_intent(user_input)

        if intent == 'greet':
            print("Chatbot: Hello! How can I assist you today?")
        elif intent == 'goodbye':
            print("Chatbot: Goodbye! Have a great day!")
            break
        elif intent == 'best_hotel':
            #city = input("In which city? ")
            match = re.search(r'in ([A-Za-z\s]+)', user_input)
            if match:
                city = match.group(1).strip()
                best_hotels = get_best_hotels_in_city(city)
                if best_hotels is not None:
                    for index, hotel in best_hotels.iterrows():
                        print(generate_detailed_response(hotel, type='hotel'))
        elif intent == 'cheapest_hotel':
            #city = input("In which city? ")
            match = re.search(r'in ([A-Za-z\s]+)', user_input)
            if match:
                city = match.group(1).strip()
                cheapest_hotel = get_cheapest_hotel_in_city(city)
                if cheapest_hotel is not None:
                    for index, hotel in cheapest_hotels.iterrows():
                        print(generate_detailed_response(cheapest_hotel, type='hotel'))
        elif intent == 'best_attraction':
            #city = input("In which city? ")
            match = re.search(r'in ([A-Za-z\s]+)', user_input)
            if match:
                city = match.group(1).strip()
                best_attraction = get_best_attractions_in_city(city)
                if best_attraction is not None:
                    for index, attraction in best_attraction.iterrows():
                        print(generate_detailed_response(attraction, type='attraction'))  
        elif intent == 'best_restaurant':
            #city = input("In which city? ")
            match = re.search(r'in ([A-Za-z\s]+)', user_input)
            if match:
                city = match.group(1).strip()
                best_restaurant = get_best_restaurant_in_city(city)
                if best_restaurant is not None:
                    for index, restaurant in best_restaurant.iterrows():
                        print(generate_detailed_response(restaurant, type='restaurant'))


        
        elif intent == 'itinerary':
            budget=0
            location = input("In which location would you like to plan an itinerary? ")
            budget=input("Enter the budget for the trip ")
            
            best_hotels, best_restaurants, best_attractions = create_itinerary_for_city(location,budget)
            print(f"Your itinerary has been planned:")
            print('*************  HOTELS ***************************')
            if best_hotels is not None:
                    for index, hotel in best_hotels.iterrows():
                        print(generate_detailed_response(hotel, type='hotel'))
            else:
                print('NO HOTEL DATA AVALIBLE FOR ', location)
            print('************* RESTAURANTS ***************************')
            if best_restaurants is not None:
                    for index, restaurant in best_restaurants.iterrows():
                        print(generate_detailed_response(restaurant, type='restaurant'))
            else:
                 print('NO RESTAURANTS DATA AVALIBLE FOR ', location)
            print('*************  TOURIST ATTRACTION   ***************************')
            if best_attractions is not None:
                    for index, attraction in best_attractions.iterrows():
                        print(generate_detailed_response(attraction, type='attraction'))
            else:
                 print('NO ATTRACTIONS DATA AVALIBLE FOR ', location)
            
            #print_itinerary(user_itinerary)
        elif intent == 'update_itinerary':
            new_activity = input("What activity would you like to add to your itinerary? ")
            user_itinerary = update_itinerary(user_itinerary, new_activity)
            print(f"Updated itinerary: {user_itinerary}")
        elif intent == 'delete_from_itinerary':
            activity = input("Which activity would you like to delete? ")
            user_itinerary = delete_activity_from_itinerary(user_itinerary, activity)
            print(f"Updated itinerary after deletion: {user_itinerary}")
        else:
            print("Chatbot: I'm sorry, I didn't understand that.")

# Start the chatbot
chatbot()
```

```python
import pandas as pd
import re
import numpy as np
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import GPT2Tokenizer, GPT2LMHeadModel, BertTokenizer, BertModel
import os

# Ensure all data files are properly loaded and handle missing data
def load_data(file_path, file_type='csv'):
    if not os.path.exists(file_path):
        print(f"Error: {file_path} not found.")
        return None
    try:
        if file_type == 'csv':
            return pd.read_csv(file_path)
        elif file_type == 'json':
            return pd.read_json(file_path)
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None

hotel_df = load_data('new_hotel_data.csv')
ratings_df = load_data('clean_hotel_review_data.json', 'json')
restaurant_df = load_data('new_restaurant_db.restaurants_data.json', 'json')
attraction_df = load_data('attraction.csv')

if hotel_df is None or ratings_df is None or restaurant_df is None or attraction_df is None:
    print("Error: One or more required datasets are missing or improperly loaded. Please check your data files.")
    exit()

# Merge hotel ratings with hotel data on 'hotel_id'
merged_df = hotel_df.merge(ratings_df, on='hotel_id', how='left')
average_ratings = merged_df.groupby('hotel_id')['rating'].mean().reset_index()
average_ratings.columns = ['hotel_id', 'average_rating']
hotel_df = hotel_df.merge(average_ratings, on='hotel_id', how='left')
hotel_df.rename(columns={'average_rating': 'Rating'}, inplace=True)

restaurant_ratings_df = load_data('new_restaurant_db.restaurants_reviews.json', 'json')
restaurant_merged_df = restaurant_df.merge(restaurant_ratings_df, on='restaurant_id', how='left')
average_restaurant_ratings = restaurant_merged_df.groupby('restaurant_id')['rating'].mean().reset_index()
average_restaurant_ratings.columns = ['restaurant_id', 'average_restaurant_ratings']
restaurant_df = restaurant_df.merge(average_restaurant_ratings, on='restaurant_id', how='left')
restaurant_df.rename(columns={'average_restaurant_ratings': 'Rating'}, inplace=True)
restaurant_df['Rating'] = restaurant_df['Rating'].fillna(3.5)

# Preprocess hotel data to generate a description
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house']

def preprocess_hotel_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."
    return description.strip()

hotel_df['description'] = hotel_df.apply(preprocess_hotel_data, axis=1)

def preprocess_restaurant_data(row):
    return f"{row['name']} located at {row['address']} and its phone number is {row['phone_number']} in {row['city']}. The provided services are {row['service']} and opening hour is {row['Open_hour']}."

def preprocess_attraction_data(row):
    return f"{row['Name']} is a popular attraction located in {row['City']}. Its address is {row['Address']}, its category is {row['Category']}, its rating is {row['Rating']}, and its availability time is {row['Availability']}. Description: {row['Description']}."

restaurant_df['description'] = restaurant_df.apply(preprocess_restaurant_data, axis=1)
attraction_df['description'] = attraction_df.apply(preprocess_attraction_data, axis=1)

# Load BERT model for embeddings
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
    return embedding

def save_embeddings(data_frame, description_column, file_name):
    if not os.path.exists(file_name):
        embedding_matrix = np.vstack([get_bert_embeddings(text) for text in data_frame[description_column]])
        np.save(file_name, embedding_matrix)
    else:
        embedding_matrix = np.load(file_name)
    return embedding_matrix

hotel_embedding_matrix = save_embeddings(hotel_df, 'description', 'hotel_embeddings.npy')
restaurant_embedding_matrix = save_embeddings(restaurant_df, 'description', 'restaurant_embeddings.npy')
attraction_embedding_matrix = save_embeddings(attraction_df, 'description', 'attraction_embeddings.npy')

def get_most_similar_entry(user_input, embedding_matrix, data_frame):
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    most_similar_index = np.argmax(similarities)
    return data_frame.iloc[most_similar_index]

# GPT-2 Model for Generative Chatbot
gpt_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
gpt_model = GPT2LMHeadModel.from_pretrained('gpt2')

def generate_response_gpt(prompt):
    inputs = gpt_tokenizer.encode(prompt, return_tensors='pt')
    outputs = gpt_model.generate(inputs, max_length=100, num_return_sequences=1)
    response = gpt_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Intents for rule-based actions
INTENTS = {
    'greet': ['hello', 'hi', 'hey', 'good morning', 'good evening'],
    'goodbye': ['goodbye', 'bye', 'see you', 'thanks', 'thank you'],
    'best_hotel': ['best hotel', 'top hotel', '5 star hotel', 'highest rated hotel'],
    'cheapest_hotel': ['cheapest hotel', 'budget hotel'],
    'best_restaurant': ['best restaurant', 'top restaurant', '5 star restaurant'],
    'best_attraction': ['best attraction', 'top attraction', 'must-see attraction'],
}

def classify_intent(user_input):
    user_input_lower = user_input.lower()
    for intent, keywords in INTENTS.items():
        for keyword in keywords:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                return intent
    return 'unknown'

def handle_intent(intent, user_input):
    if intent == 'greet':
        return "Hello! How can I assist you today?"
    elif intent == 'goodbye':
        return "Goodbye! Have a great day!"
    elif intent == 'best_hotel':
        match = re.search(r'in ([A-Za-z\s]+)', user_input)
        if match:
            city = match.group(1).strip()
            best_hotels = get_most_similar_entry(city, hotel_embedding_matrix, hotel_df)
            return generate_detailed_response(best_hotels, 'hotel')
    elif intent == 'cheapest_hotel':
        match = re.search(r'in ([A-Za-z\s]+)', user_input)
        if match:
            city = match.group(1).strip()
            cheapest_hotels = get_most_similar_entry(city, hotel_embedding_matrix, hotel_df)
            return generate_detailed_response(cheapest_hotels, 'hotel')
    elif intent == 'best_restaurant':
        match = re.search(r'in ([A-Za-z\s]+)', user_input)
        if match:
            city = match.group(1).strip()
            best_restaurant = get_most_similar_entry(city, restaurant_embedding_matrix, restaurant_df)
            return generate_detailed_response(best_restaurant, 'restaurant')
    elif intent == 'best_attraction':
        match = re.search(r'in ([A-Za-z\s]+)', user_input)
        if match:
            city = match.group(1).strip()
            best_attraction = get_most_similar_entry(city, attraction_embedding_matrix, attraction_df)
            return generate_detailed_response(best_attraction, 'attraction')
    else:
        return generate_response_gpt(user_input)

def generate_detailed_response(entry, entry_type):
    if entry_type == 'hotel':
        response = f"**{entry['name']}**\nAddress: {entry['address']}\nCity: {entry['city']}\nRating: {entry['Rating']}\nDescription: {entry['description']}"
    elif entry_type == 'restaurant':
        response = f"**{entry['name']}**\nAddress: {entry['address']}\nCity: {entry['city']}\nRating: {entry['Rating']}\nDescription: {entry['description']}"
    elif entry_type == 'attraction':
        response = f"**{entry['Name']}**\nAddress: {entry['Address']}\nCategory: {entry['Category']}\nRating: {entry['Rating']}\nDescription: {entry['Description']}"
    return response

# Main chatbot function
def chatbot():
    print("Chatbot: Hello! I can help you find the best hotels, restaurants, and attractions. How can I assist you today?")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ['exit', 'quit']:
            print("Chatbot: Goodbye! Have a great day!")
            break
        intent = classify_intent(user_input)
        response = handle_intent(intent, user_input)
        print(f"Chatbot: {response}")

# Run the chatbot
chatbot()
```

```python
import pandas as pd
import re
import numpy as np
import torch
from sklearn.metrics.pairwise import cosine_similarity
import os
from nltk.corpus import wordnet as wn
import nltk
from pymongo import MongoClient
from bson.objectid import ObjectId
import uuid  # To generate unique identifiers for trip plans
from torch.utils.data import Dataset, DataLoader
from transformers import GPT2Tokenizer, GPT2LMHeadModel, BertTokenizer, BertModel, AdamW


# Load data from files and handle missing data
def load_data(file_path, file_type='csv'):
    if not os.path.exists(file_path):
        print(f"Error: {file_path} not found.")
        return None
    try:
        if file_type == 'csv':
            return pd.read_csv(file_path)
        elif file_type == 'json':
            return pd.read_json(file_path)
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None

# Load all required datasets
hotel_df = load_data('new_hotel_data.csv')
ratings_df = load_data('clean_hotel_review_data.json', 'json')
restaurant_df = load_data('new_restaurant_db.restaurants_data.json', 'json')
attraction_df = load_data('attraction.csv')

if hotel_df is None or ratings_df is None or restaurant_df is None or attraction_df is None:
    print("Error: One or more required datasets are missing or improperly loaded. Please check your data files.")
    exit()

# Merge hotel ratings with hotel data on 'hotel_id'
merged_df = hotel_df.merge(ratings_df, on='hotel_id', how='left')
average_ratings = merged_df.groupby('hotel_id')['rating'].mean().reset_index()
average_ratings.columns = ['hotel_id', 'average_rating']
hotel_df = hotel_df.merge(average_ratings, on='hotel_id', how='left')
hotel_df.rename(columns={'average_rating': 'Rating'}, inplace=True)

restaurant_ratings_df = load_data('new_restaurant_db.restaurants_reviews.json', 'json')
restaurant_merged_df = restaurant_df.merge(restaurant_ratings_df, on='restaurant_id', how='left')
average_restaurant_ratings = restaurant_merged_df.groupby('restaurant_id')['rating'].mean().reset_index()
average_restaurant_ratings.columns = ['restaurant_id', 'average_restaurant_ratings']
restaurant_df = restaurant_df.merge(average_restaurant_ratings, on='restaurant_id', how='left')
restaurant_df.rename(columns={'average_restaurant_ratings': 'Rating'}, inplace=True)
restaurant_df['Rating'] = restaurant_df['Rating'].fillna(3.5)

# List of facilities
facilities_list = ['Cable TV', 'Entire apartment', 'Oven stove', 'Car Rental Service', 'Ironing board', 
                   'bathroom', 'Washing machine', 'Cot', 'Entire bungalow', 'Conference Room', 
                   'Fitness Center', 'Outdoor grill', 'Fitness centre', 'Pet-friendly', 'Entire house',
                   'Pet Friendly', 'Airport shuttle', 'Terrace', 'Room Service', 'BBQ Facilities', 
                   '24-Hour Front Desk', 'Child-friendly', 'Kitchen', 'Microwave', 'Free parking', 
                   'Breakfast Buffet', 'Swimming Pool', 'Paid parking', 'Entire villa', 'Air conditioning', 
                   'Balcony', 'Spa', 'Business Center', 'Free breakfast', 'Currency Exchange', 
                   'Outdoor pool', 'Air Conditioning', 'Laundry', 'Golf Course', 'cafe', 'Indoor pool', 
                   'Parking', 'Garden', 'Patio', 'Fireplace', 'Restaurant', 'Smoking Rooms', 
                   'Babysitting Service', 'Heating', 'Paid Wi-Fi', 'Wheelchair accessible', 'Entire chalet', 
                   'Lift', 'Smoke-free', 'Entire cottage', 'Library', 'Hot tub', 'Beach access', 
                   'Family Rooms', 'Free Wi-Fi']
# Initialize MongoDB client
mongo_client = MongoClient('mongodb://localhost:27017/')
db = mongo_client['travel_planner']  # Create or use the 'travel_planner' database
trip_plans_collection = db['trip_plans']  # Create or use the 'trip_plans' collection


def preprocess_hotel_data(row):
    description = f"{row['name']} and its address is {row['address']} in {row['city']}, costing Rs{row['price']} per night. Small description: {row['about']}."
    available_facilities = [facility for facility in facilities_list if row.get(facility, 0) == 1]
    if available_facilities:
        description += " Available facilities: " + ", ".join(available_facilities) + "."
    else:
        description += " No special facilities are listed."
    return description.strip()

hotel_df['description'] = hotel_df.apply(preprocess_hotel_data, axis=1)

def preprocess_restaurant_data(row):
    return f"{row['name']} located at {row['address']} and its phone number is {row['phone_number']} in {row['city']}. The provided services are {row['service']} and opening hour is {row['Open_hour']}."

def preprocess_attraction_data(row):
    return f"{row['Name']} is a popular attraction located in {row['city']}. Its address is {row['Address']}, its category is {row['Category']}, its rating is {row['Rating']}, and its availability time is {row['Availability']}. Description: {row['Description']}."

restaurant_df['description'] = restaurant_df.apply(preprocess_restaurant_data, axis=1)
attraction_df['description'] = attraction_df.apply(preprocess_attraction_data, axis=1)

# Load BERT model for embeddings
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()

def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
    return embedding

def save_embeddings(data_frame, description_column, file_name):
    if not os.path.exists(file_name):
        embedding_matrix = np.vstack([get_bert_embeddings(text) for text in data_frame[description_column]])
        np.save(file_name, embedding_matrix)
    else:
        embedding_matrix = np.load(file_name)
    return embedding_matrix

hotel_embedding_matrix = save_embeddings(hotel_df, 'description', 'hotel_embeddings.npy')
restaurant_embedding_matrix = save_embeddings(restaurant_df, 'description', 'restaurant_embeddings.npy')
attraction_embedding_matrix = save_embeddings(attraction_df, 'description', 'attraction_embeddings.npy')

# GPT-2 Model for Generative Chatbot
gpt_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
gpt_model = GPT2LMHeadModel.from_pretrained('gpt2')

# Set the padding token for the GPT-2 tokenizer
gpt_tokenizer.pad_token = gpt_tokenizer.eos_token  # Set padding token to EOS token

class GPT2BERTDataset(Dataset):
    def __init__(self, descriptions, embeddings, tokenizer):
        self.descriptions = descriptions
        self.embeddings = embeddings
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.descriptions)

    def __getitem__(self, idx):
        # Convert text description to tokens
        description = self.descriptions[idx]
        embedding = self.embeddings[idx]
        
        # Tokenize description with padding
        tokenized_input = self.tokenizer.encode(description, return_tensors='pt', padding='max_length', max_length=100, truncation=True).squeeze(0)
        
        return tokenized_input, torch.tensor(embedding, dtype=torch.float32)

# Prepare dataset and dataloader
descriptions = hotel_df['description'].tolist()  # Example using hotel descriptions
embeddings = hotel_embedding_matrix  # Use the BERT embeddings
dataset = GPT2BERTDataset(descriptions, embeddings, gpt_tokenizer)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Fine-tuning GPT-2 with BERT embeddings
optimizer = AdamW(gpt_model.parameters(), lr=5e-5)

# Training loop
for epoch in range(15):  # Adjust number of epochs as necessary
    gpt_model.train()
    for batch in dataloader:
        inputs, bert_embeddings = batch
        
        # Move tensors to the appropriate device (GPU if available)
        inputs = inputs.to(gpt_model.device)
        bert_embeddings = bert_embeddings.to(gpt_model.device)

        # Forward pass through GPT-2
        outputs = gpt_model(inputs, labels=inputs)
        
        # Calculate loss and backpropagate
        loss = outputs.loss
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch + 1} completed. Loss: {loss.item()}")

# Save the fine-tuned GPT-2 model and tokenizer
gpt_model.save_pretrained('fine_tuned_gpt2')
gpt_tokenizer.save_pretrained('fine_tuned_gpt2')


def generate_response_gpt(prompt):
    inputs = gpt_tokenizer.encode(prompt, return_tensors='pt')
    with torch.no_grad():
        # Adjust the model to take embeddings into account if modified
        outputs = gpt_model.generate(inputs, max_length=100, num_return_sequences=1, temperature=0.7, top_p=0.9)
    response = gpt_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Function to get synonyms for a word
def get_synonyms(word):
    synonyms = set()
    for synset in wn.synsets(word):
        for lemma in synset.lemmas():
            synonyms.add(lemma.name().lower())
    return synonyms

# Intents for rule-based actions with extended keywords
INTENTS = {
    'greet': ['hello', 'hi', 'hey', 'good morning', 'good evening'],
    'goodbye': ['goodbye', 'bye', 'see you', 'thanks', 'thank you'],
    'hotel': ['hotel', 'hotels', 'place to stay', 'accommodation'],
    'restaurant': ['restaurant', 'restaurants', 'place to eat', 'dining', 'food'],
    'attraction': ['attraction', 'attractions', 'thing to do', 'sightseeing', 'tourist spot'],
    'recommend': ['recommend', 'suggest', 'advise', 'best', 'top'],
    'find': ['find', 'search', 'show', 'looking for', 'need'],
    'plan': ['itinerary', 'plan', 'schedule', 'trip'],
    'update': ['update', 'change', 'modify', 'revise', 'alter']  
}




def extract_city(user_input):
   
    # Pattern to catch city names after "in", "at", "near", "around"
    match = re.search(r'\b(in|at|near|around)\s+([A-Za-z\s]+)', user_input)
    if match:
        
        return match.group(2).strip()
    
    # Split user input and check for city names
    city_names = set(hotel_df['city'].str.lower()).union(set(restaurant_df['city'].str.lower()), set(attraction_df['city'].str.lower()))
    pieces = user_input.split()
    for piece in pieces:
        if piece.lower() in city_names:
            
            return piece.strip()
    
    return None



def classify_intent(user_input):
    user_input_lower = user_input.lower()
    
    # Check direct intent match
    matched_intents = set()
    for intent, keywords in INTENTS.items():
        for keyword in keywords:
            if re.search(r'\b' + re.escape(keyword) + r'\b', user_input_lower):
                matched_intents.add(intent)
    if not matched_intents:
        user_words = user_input_lower.split()
        for intent, keywords in INTENTS.items():
            for keyword in keywords:
                synonyms = get_synonyms(keyword)
                if any(word in synonyms for word in user_words):
                    matched_intents.add(intent)
    # If no direct match, split user input and check for specific keywords
    if not matched_intents:
        pieces = user_input_lower.split()
        
        # Check for hotels, restaurants, and attractions directly in the pieces
        for piece in pieces:
            if piece in ['hotel', 'hotels']:
                matched_intents.add('hotel')
                break  # No need to check further if one is found
            elif piece in ['restaurant', 'restaurants']:
                matched_intents.add('restaurant')
                break
            elif piece in ['attraction', 'attractions']:
                matched_intents.add('attraction')
                break

    return matched_intents if matched_intents else {'unknown'}

def save_trip_plan_to_mongodb(trip_plan):
    # Assign a unique identifier (e.g., ObjectId) to each trip plan
    trip_plan['_id'] = trip_plans_collection.count_documents({}) + 1  # Unique ID
    trip_plans_collection.insert_one(trip_plan)


# Function to get the top N similar entries with city filtering
def get_top_similar_entries(user_input, embedding_matrix, data_frame, top_n=1, city=None):
    user_input_embedding = get_bert_embeddings(user_input).reshape(1, -1)
    similarities = cosine_similarity(user_input_embedding, embedding_matrix)
    
    # Filter entries by city if specified
    if city:
        filtered_df = data_frame[data_frame['city'].str.lower() == city.lower()]
        if filtered_df.empty:
            return None
        filtered_indices = filtered_df.index.values
        similarities = similarities[0][filtered_indices]  # Only compute similarities for filtered entries
        top_indices = np.argsort(similarities)[-top_n:][::-1]
        return filtered_df.iloc[top_indices]
    else:
        # No city specified, return top N overall
        top_indices = np.argsort(similarities[0])[-top_n:][::-1]
        return data_frame.iloc[top_indices]
        
def distribute_over_days(recommendations, num_days, items_per_day=3):
    recommendations_per_day = []
    for day in range(num_days):
        start_idx = (day * items_per_day) % len(recommendations)
        end_idx = start_idx + items_per_day
        recommendations_per_day.append(recommendations[start_idx:end_idx])
    return recommendations_per_day

def update_trip_plan(user_input, user_id):
    # Extract the user's trip plan from MongoDB
    trip_plan = get_trip_plan_from_mongodb(user_id)

    if not trip_plan:
        return "No trip plan found for your user ID."

    # Identify the day and the type of update requested
    day_number = extract_day_number(user_input)  # Function to extract day number from user input
    if day_number is None or day_number < 1 or day_number > trip_plan['num_days']:
        return "Please specify a valid day number for the update."

    # Get top restaurants excluding those already suggested
    existing_restaurants = set()
    for day in trip_plan['daily_plan']:
        existing_restaurants.update(restaurant['name'] for restaurant in day['restaurants'])

    top_restaurants = get_top_similar_entries(user_input, restaurant_embedding_matrix, restaurant_df, top_n=9, city=trip_plan['city'])
    
    # Filter out already suggested restaurants
    new_restaurants = top_restaurants[~top_restaurants['name'].isin(existing_restaurants)]

    # Check if we have enough new restaurant options
    if new_restaurants.shape[0] < 3:
        return "Not enough new restaurant options available to suggest."

    # Suggest three new restaurants for the specified day
    suggested_restaurants = new_restaurants.sample(n=3).to_dict('records')

    # Update the trip plan for the specified day
    trip_plan['daily_plan'][day_number - 1]['restaurants'] = [
        {'name': restaurant['name'], 'address': restaurant['address'], 'rating': restaurant['Rating']} 
        for restaurant in suggested_restaurants
    ]

    # Save the updated trip plan to MongoDB
    save_trip_plan_to_mongodb(trip_plan)

    # Create response
    responses = []
    responses.append(f"Your trip plan has been updated for Day {day_number}. Here are the new restaurant suggestions:")
    for restaurant in suggested_restaurants:
        responses.append(f"- {restaurant['name']}, {restaurant['address']} (Rating: {restaurant['Rating']})")

    return "\n".join(responses)

def get_trip_plan_from_mongodb(user_id):
    return trip_plans_collection.find_one({'user_id': user_id})

def extract_day_number(user_input):
    
    match = re.search(r'\bday\s*(\d+)\b', user_input, re.IGNORECASE)
    return int(match.group(1)) if match else None

def save_trip_plan_to_mongodb(trip_plan):
    # Ensure the trip_plan has an _id, if not, assign a new ObjectId
    if '_id' not in trip_plan:
        trip_plan['_id'] = ObjectId()  # Generate a new ObjectId

    # Insert the trip plan into MongoDB
    trip_plans_collection.insert_one(trip_plan)  # Use insert_one instead of update_one

# Modify handle_intent to generate detailed responses
def handle_intent(matched_intents, user_input):
    city = extract_city(user_input)
    responses = []

    if 'greet' in matched_intents:
        responses.append("Hello! How can I assist you today?")

    
    if 'goodbye' in matched_intents:
        responses.append("Goodbye! Have a great day.")
    
        
    if 'plan' in matched_intents and city:
        try:
            num_days = int(input("How many days are you planning to stay? "))
        except ValueError:
            responses.append("Please provide a valid number of days.")
            return "\n".join(responses)
    
        top_hotels = get_top_similar_entries(user_input, hotel_embedding_matrix, hotel_df, top_n=3, city=city)
        top_restaurants = get_top_similar_entries(user_input, restaurant_embedding_matrix, restaurant_df, top_n=9, city=city)
        top_attractions = get_top_similar_entries(user_input, attraction_embedding_matrix, attraction_df, top_n=9, city=city)
    
        # Create a trip data structure
        trip_plan = {
            '_id': ObjectId(),  # Optional: Generate a new ObjectId if needed
            'city': city,
            'num_days': num_days,
            'daily_plan': []  # Change to hold daily plans
        }

    
        if top_hotels is not None and not top_hotels.empty:
            responses.append(f"Here are the top hotels in {city} for your stay:")
            trip_plan['hotels'] = []  # List of hotels for the entire stay
            for idx, row in top_hotels.iterrows():
                responses.append(generate_detailed_response(row, 'hotel'))
                trip_plan['hotels'].append({'name': row['name'], 'address': row['address'], 'rating': row['Rating']})
    
        if top_restaurants is not None and not top_restaurants.empty:
            restaurant_days = distribute_over_days(top_restaurants.to_dict('records'), num_days, items_per_day=3)
    
        if top_attractions is not None and not top_attractions.empty:
            attraction_days = distribute_over_days(top_attractions.to_dict('records'), num_days, items_per_day=3)
    
        for day in range(num_days):
            day_plan = {
                'day': day + 1,
                'restaurants': [],
                'attractions': []
            }
            
            # Add restaurants for the current day
            if top_restaurants is not None and not top_restaurants.empty:
                day_restaurants = restaurant_days[day]
                for restaurant in day_restaurants:
                    day_plan['restaurants'].append({'name': restaurant['name'], 'address': restaurant['address'], 'rating': restaurant['Rating']})
    
            # Add attractions for the current day
            if top_attractions is not None and not top_attractions.empty:
                day_attractions = attraction_days[day]
                for attraction in day_attractions:
                    day_plan['attractions'].append({'name': attraction['Name'], 'address': attraction['Address'], 'rating': attraction['Rating']})
    
            # Add the daily plan to the trip plan
            trip_plan['daily_plan'].append(day_plan)
    
        # Save the trip plan to MongoDB
        save_trip_plan_to_mongodb(trip_plan)
    
        # Print the trip plan to the user
        responses.append("Your trip plan has been saved successfully! Here is your plan:")
        for day in trip_plan['daily_plan']:
            responses.append(f"----------------- Day {day['day']} ------------------")
            
            # Print hotels
            responses.append("Hotels:")
            for hotel in trip_plan['hotels']:
                responses.append(f"- {hotel['name']}, {hotel['address']} (Rating: {hotel['rating']})")
    
            # Print restaurants for the current day
            responses.append("Restaurants:")
            for restaurant in day['restaurants']:
                responses.append(f"- {restaurant['name']}, {restaurant['address']} (Rating: {restaurant['rating']})")
    
            # Print attractions for the current day
            responses.append("Attractions:")
            for attraction in day['attractions']:
                responses.append(f"- {attraction['name']}, {attraction['address']} (Rating: {attraction['rating']})")
    
        return "\n".join(responses)
    if 'update plan' in matched_intents and city:
            # Update the trip plan based on user input
            try:
                update_response = update_trip_plan(user_input, user_id)
                responses.append(update_response)
            except Exception as e:
                responses.append(f"An error occurred while updating your plan: {str(e)}")
    if 'hotel' in matched_intents and city:
        top_hotels = get_top_similar_entries(user_input, hotel_embedding_matrix, hotel_df, top_n=5, city=city)
        
        if top_hotels is not None:
            
            if 'hotels' in user_input:
                responses.append(f"Here are the top hotels in {city}:")
                responses.append("==================HOTELS================")
                for idx, row in top_hotels.iterrows():
                    responses.append("--------------------------------------------------------------------------------")
                    responses.append(generate_detailed_response(row, 'hotel'))
            else:
                responses.append(f"Here are the top hotel in {city}:")
                responses.append("===============HOTEL=================")
                first_hotel_row = top_hotels.iloc[0]  # Get the first attraction
                responses.append("----------------------------------------------")
                responses.append(generate_detailed_response(first_hotel_row, 'hotel'))
        else:
            responses.append(f"Sorry, no hotels found in {city}.")

    if 'restaurant' in matched_intents and city:
        top_restaurants = get_top_similar_entries(user_input, restaurant_embedding_matrix, restaurant_df, top_n=5, city=city)
        if top_restaurants is not None:
            
            if 'restaurants' in user_input:
                responses.append(f"Here are the top restaurants in {city}:")
                responses.append("==================RESTAURANTS=================")
                for idx, row in top_restaurants.iterrows():
                    responses.append("----------------------------------------------")
                    responses.append(generate_detailed_response(row, 'restaurant'))
            else:
                responses.append(f"Here are the top restaurant in {city}:")
                responses.append("==================RESTAURAT====================")
                first_restaurant_row = top_restaurants.iloc[0]  # Get the first attraction
                responses.append("----------------------------------------------")
                responses.append(generate_detailed_response(first_restaurant_row, 'restaurant'))
        else:
            responses.append(f"Sorry, no restaurants found in {city}.")
    
    
    if 'attraction' in matched_intents and city:
        top_attractions = get_top_similar_entries(user_input, attraction_embedding_matrix, attraction_df, top_n=5, city=city)
    
        if top_attractions is not None:  
            if 'attractions' in user_input:
                responses.append(f"Here are the top attractions in {city}:")
                responses.append("====================ATTRACTIONS================")
                for idx, row in top_attractions.iterrows():
                    responses.append("----------------------------------------------")
                    responses.append(generate_detailed_response(row, 'attraction'))
            else:
                responses.append(f"Here are the top attraction in {city}:")
                responses.append("================ATTRACTION=================")
                first_attraction_row = top_attractions.iloc[0]  # Get the first attraction
                responses.append("----------------------------------------------")
                responses.append(generate_detailed_response(first_attraction_row, 'attraction'))
        else:
            responses.append(f"Sorry, no attractions found in {city}.")

    if not responses:
        responses.append(generate_response_gpt(user_input))

    return "\n".join(responses)

    
def generate_detailed_response(entry, entry_type):
    if entry_type == 'hotel':
        response = f"Name: {entry['name']}\nAddress: {entry['address']}\nCity: {entry['city']}\nRating: {entry['Rating']}\nDescription: {entry['description']}"
    elif entry_type == 'restaurant':
        response = f"Name :{entry['name']}\nAddress: {entry['address']}\nCity: {entry['city']}\nRating: {entry['Rating']}\nDescription: {entry['description']}"
    elif entry_type == 'attraction':
        response = f"Name :{entry['Name']}\nAddress: {entry['Address']}\nCity : {entry['city']}\nCategory: {entry['Category']}\nRating: {entry['Rating']}\nDescription: {entry['Description']}"
    return response

# Main chatbot function
def chatbot():
    print("Chatbot: Hello! I can help you find the best hotels, restaurants, and attractions. You can also ask me anything!")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ['exit', 'quit']:
            print("Chatbot: Goodbye! Have a great day!")
            break
        intent = classify_intent(user_input)
        response = handle_intent(intent, user_input)
        print(f"Chatbot: {response}")

# Run the chatbot
chatbot()
```
